```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
response=requests.get("https://www.imdb.com/search/title?release_date=2019&sort=num_votes,desc&page=1&ref_=adv_nxt")
html_data=response.text
data=BeautifulSoup(html_data, "html.parser")
data=data.find(class_="lister-list")
detailed_list=[]
for d in data.find_all(class_="lister-item mode-advanced"):
    title=d.find(class_="lister-item-header").a.string
    year=d.find(class_="lister-item-year text-muted unbold").string[1:-1]
    runtime=float(d.find(class_="runtime").string[:3])
    rating=float(d.find(class_="inline-block ratings-imdb-rating").strong.string[:3])
    votes=float(list(d.find(class_="sort-num_votes-visible").strings)[3].replace(",",""))
    try:
        gross=list(d.find(class_="sort-num_votes-visible").strings)[9]
    except:
        continue
    genre=d.find(class_="genre").string.strip()
    detailed_list.append([title,year,genre,runtime,rating,votes,gross])
    
MovieData = pd.DataFrame(detailed_list, columns=["Title","Year","Genre","Runtime","Rating","Votes","Gross"])
```
